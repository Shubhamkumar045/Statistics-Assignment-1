```python
Q1. What are the three measures of central tendency?
Answer--The three measures of central tendency are:

Mean: The mean, also known as the average, is calculated by summing up all the values 
in a dataset and dividing by the total number of values. It represents the arithmetic average of the dataset.

Median: The median is the middle value of a dataset when the values are arranged in
ascending or descending order. If there is an even number of values, the median is
the average of the two middle values. The median represents the middle value of the dataset.

Mode: The mode is the value that appears most frequently in a dataset. A dataset
can have one mode (unimodal), two modes (bimodal), or more than two modes (multimodal).
If no value is repeated, the dataset is said to have no mode. The mode represents the 
most common value(s) in the dataset.

Q2. What is the difference between the mean, median, and mode? How are they used to measure the
central tendency of a dataset?
Answer--
Mean:

Definition: The mean, also known as the arithmetic average, is calculated by summing up all the values in
a dataset and then dividing by the total number of values.
Calculation: Mean = (Sum of all values) / (Number of values)
Usage: The mean is sensitive to extreme values, making it a good measure of central tendency for
datasets with a symmetric distribution and no outliers. It provides a balanced representation of the dataset's values.
Median:

Definition: The median is the middle value of a dataset when the values are arranged in ascending or 
descending order. If there is an even number of values, the median is the average of the two middle values.
Usage: The median is robust to outliers and is often used for datasets with skewed distributions or
when extreme values are present. It represents the value below and above which 50% of the values lie.
Mode:

Definition: The mode is the value that appears most frequently in a dataset.
Usage: The mode is useful for identifying the most common value or category in a dataset. 
It is often used for categorical or discrete data and can be used alongside other measures of central tendency.
Differences:

The mean is influenced by extreme values, while the median is not affected by outliers.
The mode is the only measure that can be applied to categorical data, while the mean and
median are typically used for numerical data.
The mean is a precise measure of central tendency, while the median and mode provide robust
measures, especially in the presence of outliers or skewed distributions.

Q3. Measure the three measures of central tendency for the given height data:
[178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]
Answer--Mean:

Mean = (Sum of all values) / (Number of values)
Mean = (178 + 177 + 176 + 177 + 178.2 + 178 + 175 + 179 + 180 + 175 + 178.9 + 176.2 + 177 + 172.5 + 178 + 176.5) / 16
Mean ≈ 177.80625
Median:

First, sort the data in ascending order: [172.5, 175, 175, 176, 176, 176.2, 176.5, 177, 177, 177, 178, 178, 178, 178.2, 178.9, 180]
Since there are 16 values, the median is the average of the two middle values: (177 + 178) / 2 = 177.5
Mode:

The mode is the value that appears most frequently in the dataset.
In the given dataset, the mode is 178, as it appears three times, more than any other value.
So, for the given height data:

Mean ≈ 177.81
Median = 177.5
Mode = 178

Q4. Find the standard deviation for the given data:
[178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]
Answer--Calculate the mean (average) of the data:

Mean
=
Sum of all data points
Number of data points
Mean= 
Number of data points
Sum of all data points
​
 
Mean
=
178
+
177
+
176
+
177
+
178.2
+
178
+
175
+
179
+
180
+
175
+
178.9
+
176.2
+
177
+
172.5
+
178
+
176.5
16
Mean= 
16
178+177+176+177+178.2+178+175+179+180+175+178.9+176.2+177+172.5+178+176.5
​
 
Mean
≈
177.80625
Mean≈177.80625
Calculate the squared difference from the mean for each data point:

Subtract the mean from each data point and square the result.
Calculate the variance:

Average the squared differences.
Calculate the standard deviation:

Take the square root of the variance.
Let's calculate it step by step:

Calculate the mean:

Mean
≈
177.80625
Mean≈177.80625
Calculate the squared difference from the mean for each data point:

(
178
−
177.80625
)
2
(
177
−
177.80625
)
2
(
176
−
177.80625
)
2
⋮
(
176.5
−
177.80625
)
2
​
  
(178−177.80625) 
2
 
(177−177.80625) 
2
 
(176−177.80625) 
2
 
⋮
(176.5−177.80625) 
2
 
​
 
Calculate the variance:

Variance
=
Sum of squared differences
Number of data points
Variance= 
Number of data points
Sum of squared differences
​
 
Calculate the standard deviation:

Standard Deviation
=
Variance
Standard Deviation= 
Variance
​
 
Let's perform the calculations to find the standard deviation.

Given data: [178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5]

Calculate the mean:

Mean
=
178
+
177
+
176
+
177
+
178.2
+
178
+
175
+
179
+
180
+
175
+
178.9
+
176.2
+
177
+
172.5
+
178
+
176.5
16
Mean= 
16
178+177+176+177+178.2+178+175+179+180+175+178.9+176.2+177+172.5+178+176.5
​
 
Mean
≈
177.80625
Mean≈177.80625
Calculate the squared difference from the mean for each data point:

(
178
−
177.80625
)
2
≈
0.033953
(
177
−
177.80625
)
2
≈
0.665352
(
176
−
177.80625
)
2
≈
3.315103
(
177
−
177.80625
)
2
≈
0.665352
⋮
(
176.5
−
177.80625
)
2
≈
2.109103
​
  
(178−177.80625) 
2
 ≈0.033953
(177−177.80625) 
2
 ≈0.665352
(176−177.80625) 
2
 ≈3.315103
(177−177.80625) 
2
 ≈0.665352
⋮
(176.5−177.80625) 
2
 ≈2.109103
​
 
Calculate the variance:

Variance
=
0.033953
+
0.665352
+
…
+
2.109103
16
Variance= 
16
0.033953+0.665352+…+2.109103
​
 
Variance
≈
2.178072
Variance≈2.178072
Calculate the standard deviation:

Standard Deviation
=
2.178072
Standard Deviation= 
2.178072
​
 
Standard Deviation
≈
1.475839
Standard Deviation≈1.475839

Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe
the spread of a dataset? Provide an example.
Answer--Measures of dispersion, including range, variance, and standard deviation, are used to describe the spread or variability of a dataset. Here's how each of these measures contributes to understanding the spread of a dataset:

Range:

The range is the simplest measure of dispersion and represents the difference between the maximum and minimum values in a dataset.
It provides a basic understanding of how spread out the data points are.
However, it can be influenced greatly by outliers and may not provide a complete picture of variability.
Variance:

Variance measures how much the values in a dataset deviate from the mean.
It calculates the average of the squared differences between each data point and the mean.
Variance is useful for understanding the overall variability of a dataset.
However, it is not in the same units as the original data, making it less intuitive to interpret.
Standard Deviation:

Standard deviation is the square root of the variance.
It measures the average deviation of data points from the mean.
It provides a measure of dispersion in the same units as the original data, making it more interpretable.
Standard deviation is widely used and provides a more precise measure of variability compared to the range.
Example:
Consider the following dataset representing the daily temperatures (in degrees Celsius) recorded over a week:
{
20
,
22
,
19
,
23
,
21
,
20
,
25
}
{20,22,19,23,21,20,25}

Q6. What is a Venn diagram?
Answer--A Venn diagram is a graphical representation used to illustrate the relationships between 
different sets or groups of items. It consists of overlapping circles, each representing a set or
group, with the overlapping areas indicating the intersections or commonalities between the sets.

Key features of Venn diagrams include:

Circles or Ellipses: Each circle or ellipse in a Venn diagram represents a set or group of items.
The items within each set share common characteristics or attributes.

Overlap: The overlapping areas between circles represent the intersections or common elements shared
by the sets. The size of the overlap indicates the extent of shared elements between the sets.

Regions: The non-overlapping regions within each circle represent the unique elements specific to that set.

Venn diagrams are commonly used in various fields, including mathematics, statistics, logic, and data
visualization, to visually depict relationships between different categories or entities. 

Q7. For the two given sets A = (2,3,4,5,6,7) & B = (0,2,6,8,10). Find:
(i) A B
(ii) A ⋃ B
Answer--To find the operations 
�
∩
�
A∩B (intersection) and 
�
∪
�
A∪B (union) for the given sets 
�
A and 
�
B, we need to understand the basic set operations:

(i) Intersection (
�
∩
�
A∩B): This operation yields a set containing all elements that are common to both sets 
�
A and 
�
B.

(ii) Union (
�
∪
�
A∪B): This operation results in a set containing all distinct elements from both sets 
�
A and 
�
B, with no repetition.

Given sets:
�
=
{
2
,
3
,
4
,
5
,
6
,
7
}
A={2,3,4,5,6,7}
�
=
{
0
,
2
,
6
,
8
,
10
}
B={0,2,6,8,10}

(i) Intersection (
�
∩
�
A∩B):
To find 
�
∩
�
A∩B, we identify the elements common to both sets 
�
A and 
�
B:
�
∩
�
=
{
2
,
6
}
A∩B={2,6}

(ii) Union (
�
∪
�
A∪B):
To find 
�
∪
�
A∪B, we combine all the distinct elements from sets 
�
A and 
�
B:
�
∪
�
=
{
0
,
2
,
3
,
4
,
5
,
6
,
7
,
8
,
10
}
A∪B={0,2,3,4,5,6,7,8,10}

Therefore:
(i) 
�
∩
�
=
{
2
,
6
}
A∩B={2,6}
(ii) 
�
∪
�
=
{
0
,
2
,
3
,
4
,
5
,
6
,
7
,
8
,
10
}
A∪B={0,2,3,4,5,6,7,8,10}

Q8. What do you understand about skewness in data?
Answer--
Skewness in data is a measure of the asymmetry of the distribution of values in a dataset. 
It indicates the degree to which the data deviates from symmetry around its mean. In other words, skewness measures the lack of symmetry in the distribution of data points.

Key points about skewness in data include:

Symmetry: A symmetric distribution has data points evenly distributed around the mean, with equal tails on both sides of the distribution. In a symmetric distribution, the mean, median, and mode are approximately equal.

Skewed Distribution: Skewness occurs when the distribution of data is not symmetrical. It can be either positively skewed (skewed to the right) or negatively skewed (skewed to the left).

Positive Skewness (Right Skew): In a positively skewed distribution, the tail of the distribution extends towards the higher values, and the mean is typically greater than the median. The majority of data points are concentrated on the left side of the distribution.

Negative Skewness (Left Skew): In a negatively skewed distribution, the tail of the distribution extends towards the lower values, and the mean is typically less than the median. The majority of data points are concentrated on the right side of the distribution.

Skewness Coefficient: Skewness is quantified using a skewness coefficient, often denoted by 
�
1
γ 
1
​
 . This coefficient measures the degree and direction of skewness in the data distribution.

�
1
>
0
γ 
1
​
 >0: Positive skewness (right skew)
�
1
<
0
γ 
1
​
 <0: Negative skewness (left skew)
�
1
=
0
γ 
1
​
 =0: Symmetric distribution
Causes of Skewness: Skewness can occur due to various factors, such as outliers, censoring, or the nature of the underlying data-generating process. For example, financial data often exhibits positive skewness due to the presence of extreme positive values (e.g., stock returns).
```
